```python
def check(df):
    
    print(f"Shape: {df.shape}")
    
    if wf.getSevTN(df).shape[0] == 0:
        print("NO [Sevierville, TN]")
    if wf.getSevTN(df).shape[0] > 0:
        print("YES [Sevierville, TN]")
    
    print(f"PercNull: {df.isna().sum().sum()/df.size}")
    
    print(f"{df.columns[10]} to {df.columns[-1]}")
```

```python
ZHVI_All = wf.MetricMeans(df1, 'ZHVI_All')
BS_IndexG = wf.MetricMeans(df4, 'BS_IndexG')
BS_IndexT = wf.MetricMeans(df5, 'BS_IndexT')
DaysZill = wf.MetricMeans(df6, 'DaysZill')
PriceCut = wf.MetricMeans(df7, 'PriceCut')
ZHVI_sfr = wf.MetricMeans(df10, 'ZHVI_sfr')
mListing = wf.MetricMeans(df13, 'mListing')
SaleCut = wf.MetricMeans(df13, 'SaleCut')
SalePrice = wf.MetricMeans(df13, 'SalePrice')
```

```python
dfs = [ZHVI_All, BS_IndexG, BS_IndexT, DaysZill, PriceCut, ZHVI_sfr, mListing, SaleCut, SalePrice]
```

```python
df_MetricMeans = reduce(lambda left, right: pd.merge(left, right, on='RegionID', how='outer'), dfs)
```

```python
df_MetricMeans.shape
```

```python
df_MetricMeans.head()
```

```python
df13.head(2)
```

```python
descriptiveCols = df1[df1.columns[0:2]]
descriptiveCols_Listing = df13[df13.columns[0:2]]
```

```python
almost = pd.merge(descriptiveCols, df_MetricMeans, on='RegionID', how='outer')
```

```python
final = pd.merge(descriptiveCols_Listing, almost, on='RegionID', how='outer')
```

```python
final = final.drop(columns=['SizeRank_x', 'RegionName_x'])
```
